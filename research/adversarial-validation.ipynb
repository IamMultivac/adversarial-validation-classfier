# Rationale:

We are to test if by removing features that present drift according to the adversarial classifer, we are to improve the model performance on the holdout set.

# Methodology:

* We are to use the ```AdversarialValidatio``` class to train an adversarial classfier.
* SHAP values for the model will be checked to removed the top n features.
* By using sequential elimination, we are to keep only those features so that the adversarial model has a performance close to ```.7 - .6 ROC AUC```
* We shall retrain the original model without the discarted features and evaluate to see how the performance improves.
* Finally, we will verify if the model with no drift features performs better than the baseline model.

# Conclusions:

* The model performance decreased significantly on the model with no drift features.
* Eventhough the adversarial validation is aa intuitive approach to multivariate features drift, it does not guarantee that the generalization power of the model will increase.

# Improvements and furter work:

* Work in feature engineering to increase both models' predictive power.
* Plot to compare feature removal aginst performance on the original model.


````python
%load_ext autoreload
%autoreload 2

import warnings;warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid")

import pandas as pd
import numpy as np

pd.set_option("display.max_columns",None)

import opendatasets as od

import os
os.chdir("../code")

from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier

from adversarial_validation import AdversarialValidation
````

## Load the data
We are to use [IEEE-CIS Fraud Detection data](https://www.kaggle.com/competitions/ieee-fraud-detection/data) data from kaggle for the experiment.

We are to keep only meaningful features for prediction i.e droping id's and time-like columns, since as per data documentation they are no actual time columns.

````python
# os.chdir("../data")
path = "https://www.kaggle.com/competitions/ieee-fraud-detection/data"
od.download(path,force=False, data_dir='../data')
````

````python
train_df = pd.read_csv("../data/ieee-fraud-detection/train_transaction.csv")
test_df = pd.read_csv("../data/ieee-fraud-detection/test_transaction.csv")
````

````python
train_df.head()
````

````python
features_list = train_df\
                .select_dtypes(exclude = "object")\
                .drop(["isFraud","TransactionID", "TransactionDT"], axis = 1)\
                .columns\
                .tolist()

target_column = "isFraud"
````

### Train a base model with all the features.

By training an ensemble model with no hyperparameter tuning we are to stablish a benchmark of performance using KFold- CV.

We are to estimate the cv performance and the score on the test set on Kaggle.

````python
clf = XGBClassifier(random_state = 42).fit(X = train_df[features_list], y = train_df[target_column])
cv_performance = cross_val_score(estimator = clf, X = train_df[features_list], y = train_df[target_column], cv = 5, scoring="roc_auc")
````

````python
pd.Series(cv_performance).describe()
````

````python
test_df.loc[:,"prediction"] = clf.predict_proba(test_df[features_list])[:,1]
````

````python
base_model_predictions = test_df[["TransactionID","prediction"]].copy()
base_model_predictions.rename(columns = {"prediction":"isFraud"}, inplace = True)
````

````python
base_model_predictions.to_csv("../submissions/base_model_predictions.csv", index = False)
````

### Train an adversarial model.

We are to train an adversarial classifier and analyse its performance.

Since we do not have a time column in this case we need to concatenate both dataframes (Train and Test) and we need also remove the original target column.

````python
%%time
train_df.loc[:,"is_train"] = 0
test_df.loc[:,"is_train"] = 1
frame = pd.concat([train_df, test_df], ignore_index=True).sample(frac = .01)
estimator_params = {"random_state":42}
adversarial_clf = AdversarialValidation(features = features_list,
                                        is_train_column = "is_train"
                                       ).fit(frame,estimator_params)
````

````python
adversarial_clf.performance_
````

````python
adversarial_clf.plot_shap_values(plot_type="bar")
````

````python
adversarial_clf.save_model()
````

````python
removal_frame = adversarial_clf.recursive_feature_elimination(frame, n_features_remove = 10)
````

````python
fig, ax = plt.subplots(figsize = (10,4))
removal_frame.set_index("columns").plot.bar(edgecolor = "black", ax = ax)
ax.set_title("Performance of the adversarial model after feature removal")
ax.set_ylabel("ROC AUC");
````

````python
removal_frame = adversarial_clf.recursive_feature_elimination(frame, threshold_remove_until=.5)
````

````python
fig, ax = plt.subplots(figsize = (10,4))
n_removed = removal_frame.shape[0]
removal_frame.set_index("columns").plot(ax = ax)
ax.set_title(f"Performance of the adversarial model after feature removal [{n_removed} removed to .5]")
ax.set_ylabel("ROC AUC");
````

### Train a model excluding the features that make the adversarial model reach a performance of 0.65

We are to remove features so that the adversarial model loses predictive power, we would like to make it ner .5 but if so we wuold need to remove 319 out 370 features.

With a threshold of 0.65 we would keep 297 features.

````python
feats_to_remove = removal_frame[removal_frame.adversarial_model_performance > .65]["columns"].tolist()
````

````python
features_list_no_drfit = [x for x in features_list if x not in feats_to_remove]
````

````python
clf = XGBClassifier(random_state = 42).fit(X = train_df[features_list_no_drfit], y = train_df[target_column])
cv_performance = cross_val_score(estimator = clf, X = train_df[features_list_no_drfit], y = train_df[target_column], cv = 5, scoring="roc_auc")
````

````python
pd.Series(cv_performance).describe()
````

````python
test_df.loc[:,"prediction"] = clf.predict_proba(test_df[features_list_no_drfit])[:,1]
````

````python
no_drift_model_predictions = test_df[["TransactionID","prediction"]].copy()
no_drift_model_predictions.rename(columns = {"prediction":"isFraud"}, inplace = True)
````
